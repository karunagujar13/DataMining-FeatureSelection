### a. Signal-to-noise ratio 

```python
def calculate_SNR(m_a, m_b, m_c, sd_a, sd_b, sd_c):
    return ((abs(m_a - m_b - m_c)) / (sd_a + sd_b + sd_c))
```

```python
SNR_s_len = calculate_SNR(m_setosa_s_len, m_versi_s_len, m_vg_s_len, sd_setosa_s_len, sd_versi_s_len
                           , sd_vg_s_len)
SNR_s_len
```

```python
SNR_s_width = calculate_SNR(m_setosa_s_width, m_versi_s_width, m_vg_s_width, sd_setosa_s_width, sd_versi_s_width, sd_vg_s_width)
SNR_s_width
```

```python
SNR_p_len = calculate_SNR(m_setosa_p_len, m_versi_p_len, m_vg_p_len, sd_setosa_p_len, sd_versi_p_len, sd_vg_p_len)
SNR_p_len
```

```python
SNR_p_width = calculate_SNR(m_setosa_p_width, m_versi_p_width, m_vg_p_width, sd_setosa_p_width, sd_versi_p_width, 
                            sd_vg_p_width)
SNR_p_width
```

Since signal-to-ratio of "petal length" attribute is larger, "Petal length" is a good feature for classification. "Petal width" has the second largest signal-to-noise ratio. Thus petal length and width are the two best features for classification.   

### Two tailed t-test

Looking at the numbers (mean and standard deviation) above for each attribute for different classes, Virginica and Versicolor classes seems to have mean and standard deviation closer than for Virginica and Setosa. So, we need to find an attribute which is a good distinguisher between Setosa and Virginica. 


I have implemented t-test calculations and also using Scipy.stats's ttest_ind for cross checking the t-value.

Even though distinguishing Setosa from Virginica (and vice-versa) is most important I wanted to test for combinations of the classes.


```python
def calculate_t(n_a, n_b, m_a, m_b, sd_a, sd_b):
    return (((m_a - m_b )) / math.sqrt(((sd_a * sd_a)/n_a) + ((sd_b * sd_b)/n_b)))
```

#### Setosa - Virginica 

```python
degree_of_freedom = len(df_setosa) + len(df_virginica) - 2
degree_of_freedom
```

From t-table two tailed, for p=0.05 and degree_of_freedom = 95; t_critical = 1.980

```python
t_calculated = calculate_t(len(df_setosa), len(df_virginica), m_setosa_s_len, m_vg_s_len, sd_setosa_s_len, sd_vg_s_len)


print('t calculated = '+str(t_calculated))


t2, p2 = stats.ttest_ind(df_setosa['sepal_len'],df_virginica['sepal_len'],equal_var=False)
print("t from stats = " + str(t2))
```

```python
t_calculated = calculate_t(len(df_setosa), len(df_virginica), m_setosa_s_width, m_vg_s_width, sd_setosa_s_width, sd_vg_s_width)

print('t calculated = '+str(t_calculated))

t2, p2 = stats.ttest_ind(df_setosa['sepal_width'],df_virginica['sepal_width'],equal_var=False)
print("t from stats = " + str(t2))
```

```python
t_calculated = calculate_t(len(df_setosa), len(df_virginica), m_setosa_p_len, m_vg_p_len, sd_setosa_p_len, sd_vg_p_len)

print('t calculated = '+str(t_calculated))

t2, p2 = stats.ttest_ind(df_setosa['petal_len'],df_virginica['petal_len'],equal_var=False)
print("t from stats= " + str(t2))
```

```python
t_calculated = calculate_t(len(df_setosa), len(df_virginica), m_setosa_p_width, m_vg_p_width, sd_setosa_p_width, sd_vg_p_width)

print('t calculated = '+str(t_calculated))

t2, p2 = stats.ttest_ind(df_setosa['petal_width'],df_virginica['petal_width'],equal_var=False)
print("t from stats= " + str(t2))
```

Between class Setosa and Virginica, null hypothesis is that the classes are normally distributed and have equal variance. For all the attributes except Sepal width the hypothesis holds true. ### Relief

```python
def calculate_distance(obj1, obj2):
    sum = 0
    for i in range(0, 4):
        sum = sum + pow((obj1[i]-obj2[i]), 2)
    return math.sqrt(sum)
```

This function calculates distance from the given object to all the other objects in the dtaabase and returns the closest object of the same class (near hit) and closest object of the different class (near miss).

```python
def process_obj(obj):
    obj_label=[]
    obj1 = obj[:-1]
    obj_label.append(obj[-1])
    
    distance = df_iris.apply(lambda x:sqrt(calculate_distance(x.values.tolist()[0:4],obj1)),axis=1)
    df_iris['distance'] = distance
    
    df_sorted = df_iris.sort_values("distance").copy().reset_index(drop=True)
    
    tmp = df_iris.groupby("label").apply(lambda x:x.sort_values("distance").head(2))
    near_hit = list(tmp[tmp.label==obj[-1]].values[1])
    
    df_near_miss = df_iris.loc[[df_iris.loc[~df_iris.label.isin(obj_label), 'distance'].idxmin()]]
    near_miss = df_near_miss.values.tolist()
    
    return near_hit, near_miss[0]
```

```python
obj1 = [4.3,3.0,1.1,0.1,'Iris-setosa']
near_hit, near_miss = process_obj(obj1)
near_hit, near_miss
```

Range for all the columns

```python
range_sepal_len = df_iris["sepal_len"].max() - df_iris["sepal_len"].min()
range_sepal_width = df_iris["sepal_width"].max() - df_iris["sepal_width"].min()

range_petal_len = df_iris["petal_len"].max() - df_iris["petal_len"].min()
range_petal_width = df_iris["petal_width"].max() - df_iris["petal_width"].min()

range_sepal_len, range_sepal_width, range_petal_len, range_petal_width
```

Update weights

```python
W1 = 0 + pow(((obj1[0]-near_hit[0])/range_sepal_len),2) + pow(((obj1[0]-near_miss[0])/range_sepal_len),2)
W2 = 0 + pow(((obj1[1]-near_hit[1])/range_sepal_width),2) + pow(((obj1[1]-near_miss[1])/range_sepal_width),2)

W3 = 0 + pow(((obj1[2]-near_hit[2])/range_petal_len),2) + pow(((obj1[2]-near_miss[2])/range_petal_len),2)
W4 = 0 + pow(((obj1[3]-near_hit[3])/range_petal_width),2) + pow(((obj1[3]-near_miss[3])/range_petal_width),2)
```
